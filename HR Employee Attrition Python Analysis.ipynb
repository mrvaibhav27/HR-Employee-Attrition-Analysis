## Data cleaning

```python
# remove un usefull  column becouse they have constant values

read=x.drop(columns=['EmployeeCount','StandardHours','Over18'])
```

```python
read
```

## Pre-processing 

#### Convert 'Attrition' to numrical values


```python
read["Attrition"] = read["Attrition"].map({"Yes": 1, "No": 0})
```

```python
read
```

## EDA (Exploratory Data Analysis):

```python
read
```

#### 1.	Business Travel vs Attrition	

```python
sns.set_theme(style='darkgrid')
sns.barplot(x=read['BusinessTravel'], y=read['Attrition'], estimator=lambda x: sum(x)/len(x) * 100)
plt.xlabel('BusinessTravel')
plt.ylabel('Attrition in percentage')
plt.title('Business travel wise attrition')
plt.show()
```

#### 2. Department vs Attrition

```python
sns.set_theme(style='darkgrid')
sns.barplot(x=read['Department'],y=read['Attrition'],estimator=lambda x: sum(x)/len(x) * 100)
plt.xlabel('Department')
plt.ylabel('Attrition in %')
plt.title("Department wise Attrition")
plt.show()
```

#### 3. Education vs Attrition

```python
sns.set_theme(style='darkgrid')
sns.barplot(x=read['Education'],y=read['Attrition'],estimator=lambda x: sum(x)/len(x) * 100)
plt.xlabel('Department')
plt.ylabel('Attrition in %')
plt.title("Education level wise Attrition")
plt.show()
```

#### 4. Job Role vs Attrition

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(12, 5))
plt.xticks(rotation=45)
sns.barplot(x=read['JobRole'],y=read['Attrition'],estimator=lambda x: sum(x)/len(x) * 100)
plt.xlabel('Department')
plt.ylabel('Attrition in %')
plt.title("Job role wise Attrition")
plt.show()
```

#### 5. Gender vs Attrition

```python
sns.set_theme(style='darkgrid')
sns.barplot(x=read['Gender'],y=read['Attrition'],estimator=lambda x: sum(x)/len(x) * 100)
plt.xlabel('Department')
plt.ylabel('Attrition in %')
plt.title("Gender wise Attrition")
plt.show()
```

#### 6.	Any other relations with Attrition

```python
#male & female Count of  attrition 
sns.countplot(x=read['Gender'],hue=read['Attrition'])
plt.show()
```

```python
# Distances from home to  business travel of attrition
sns.barplot(x='BusinessTravel', y='DistanceFromHome',hue='Attrition', data=read)
plt.show()
```

```python
# Age wise Attrition count
plt.figure(figsize=(12, 5))
plt.xticks(rotation=45)
sns.countplot(x='Age',hue='Attrition', data=read)
plt.show()
```

```python
# Age wise Attrition
plt.figure(figsize=(8, 5))
sns.stripplot(x=read['Attrition'],y=read['Age'],color='g')
plt.xlabel("Attriton")
plt.ylabel("Age")
plt.show()
```

```python
#  Marital status wise attrition Count 
sns.countplot(x=read['MaritalStatus'],hue=read['Attrition'])
plt.show()
```

```python
#  Percent salary hike wise attrition Count 
sns.countplot(x=read['PercentSalaryHike'],hue=read['Attrition'])
plt.show()
```

```python
# Years Since Last Promotion wise attrition Count 
sns.countplot(x=read['YearsSinceLastPromotion'],hue=read['Attrition'])
plt.show()
```

```python
# years at company and years since last promotion wise attrition
plt.figure(figsize=(8, 5))
sns.stripplot(x="YearsSinceLastPromotion",y="YearsAtCompany", data=read, hue="Attrition", dodge="true")
plt.show()
```

```python
#Department wise attrition Count and total empolye count 
sns.countplot(x=read['Department'],hue=read['Attrition'])
plt.show()
```

```python
# overtime wise count
sns.countplot(x=read['OverTime'],hue=read['Attrition'])
plt.show()
```
